```python
import pandas as pd
```

```python
train = pd.read_csv('../Attrition/train.csv', index_col=0)
test = pd.read_csv('../Attrition/test.csv', index_col=0)

train.head()
```

```python
train.info()
```

```python
train['Attrition'] = train['Attrition'].map(lambda x : 1 if x =='Yes' else 0)
train.head()
```

```python
train = train.drop(['EmployeeNumber','StandardHours'], axis=1)
test = test.drop(['EmployeeNumber','StandardHours'], axis=1)
train.head()
```

```python
from sklearn.preprocessing import LabelEncoder

features = [
    'Age',
    'BusinessTravel',
    'Department',
    'EducationField',
    'Gender',
    'JobRole',
    'MaritalStatus',
    'Over18',
    'OverTime'
]

for f in features:
    labelEncoder = LabelEncoder()
    train[f] = labelEncoder.fit_transform(train[f])
    test[f] = labelEncoder.transform(test[f])

train.head()
```

```python
# train = train.drop(['YearsInCurrentRole', 'PerformanceRating'], axis=1)
# test = test.drop(['YearsInCurrentRole', 'PerformanceRating'], axis=1)
```

```python
from sklearn.linear_model.logistic import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(
    train.drop('Attrition', axis=1), 
    train['Attrition'], 
    test_size=0.2, 
    random_state=42
)
```

```python
model = LogisticRegression(
    max_iter=100,
    verbose=True,
    random_state=33,
    tol=1e-4,
)
model.fit(X_train, y_train)
```

```python
predict = model.predict_proba(test)[:, 1]
```

```python
test['Attrition']=predict
# test[['Attrition']].head()
```

```python
test[['Attrition']].to_csv('./Submission.csv')
```
