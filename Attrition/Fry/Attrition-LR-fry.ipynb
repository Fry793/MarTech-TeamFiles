```python
import pandas as pd
train =pd.read_csv('train.csv',index_col=0)
test =pd.read_csv('test.csv',index_col=0)
```

```python

#train['Attrition'].value_counts()
train.head()
```

```python
train=train.drop(['EmployeeNumber','StandardHours'],axis=1)
test=test.drop(['EmployeeNumber','StandardHours'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
train['Attrition']=train['Attrition'].map(lambda x:1 if x=='Yes' else 0)
attr=['Age','BusinessTravel','Department','Education','EducationField','Gender','JobRole','MaritalStatus','Over18','OverTime']
lbe_list=[]
for feature in attr:
    lbe=LabelEncoder()
    train[feature]=lbe.fit_transform(train[feature])
    test[feature]=lbe.transform(test[feature])
```

```python
from sklearn.model_selection import  train_test_split
train_x,test_x,train_y,test_y=train_test_split(train.drop(['Attrition'],axis=1),train['Attrition'],test_size=0.2)
```

```python
#归一化：
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler(feature_range=(0, 1))
train_x = mms.fit_transform(train_x)
test_x = mms.fit_transform(test_x)
test = mms.fit_transform(test)
```

```python
from sklearn.linear_model.logistic import LogisticRegression
LR=LogisticRegression(max_iter=100, 
                           verbose=True,
                           tol=1e-4
                          )
LR.fit(train_x,train_y)
predic_y=LR.predict(test_x)

from sklearn.metrics import accuracy_score
accuracy_score(predic_y,test_y)
```

```python
pre=LR.predict_proba(test)[:,1]
```

```python
test['Attrition']=pre
test['Attrition'].to_csv('LR_Subm.csv')
```
